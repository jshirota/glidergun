```python
from glidergun import mosaic, search

items = search(
    "landsat-c2-l2",
    (-80, 44, -79, 45),
    datetime="2025-10-01/2025-10-05",
    fully_contains_search_area=False,
    cloud_cover_percent=1,
)

s1, s2, *_ = [item.download(["red", "green", "blue"], clip_to_extent=False) for item in items]

mosaic(s1, s2), mosaic(s1, s2, blend=True)
```
