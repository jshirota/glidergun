```python
from glidergun import grid

dem = grid("cop-dem-glo-90", (137.8, 34.5, 141.1, 36.8))
hillshade = dem.hillshade()

dem.save("output/dem.tif")
hillshade.save("output/hillshade.tif", "uint8")

japan = hillshade.set_nan(dem < 1).overlay(dem.set_nan(dem < 1).color("terrain").to_stack(), alpha=0.5)
japan.save("output/japan.png")

japan
```
