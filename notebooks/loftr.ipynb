```python
from glidergun import stack

esri = stack("esri", (-123.20, 49.26, -123.07, 49.32))

old_aerial_photo = stack("../data/vancouver.png")
georeferenced = old_aerial_photo.georeference_to(esri)

old_aerial_photo, georeferenced, esri, esri.overlay(georeferenced)
```

```python
from glidergun import grid, stack

google = grid("google", (-79.418, 43.648, -79.388, 43.672), max_tiles=50)
toronto = stack("https://www.toronto.ca/ext/archives/s0012/fl1947/s0012_fl1947_it0018f.jpg").mean()

georeferenced = toronto.georeference_to(google)
# georeferenced.save("output/toronto.tif", "int16")

toronto, google.overlay(georeferenced.color("Reds").to_stack(), alpha=0.7)
```
