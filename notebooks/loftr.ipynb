```python
from glidergun import stack

bing = stack("microsoft", (-123.28, 49.21, -123.07, 49.34), max_tiles=100)
vancouver = stack("../data/vancouver.png")

georeferenced = vancouver.georeference_to(bing)

georeferenced, bing.overlay(georeferenced)
```

```python
from glidergun import grid, stack

google = grid("google", (-79.430, 43.640, -79.388, 43.672), max_tiles=100)
toronto = stack("https://www.toronto.ca/ext/archives/s0012/fl1947/s0012_fl1947_it0018f.jpg")

georeferenced = toronto.georeference_to(google)

google.overlay(georeferenced, alpha=0.7)
```
