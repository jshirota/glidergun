```python
from glidergun import grid, stack

e = (-123.3, 49.0, -122.8, 49.3)

google = stack("google", e, max_tiles=50)
dem = grid("cop-dem-glo-90", e).project(3857)

google.overlay(dem.set_nan(dem > 2).color("Blues_r").to_stack(), alpha=0.7)
```

```python
from IPython.display import Markdown

from glidergun import animate, stack

dates = [f"2026-01-{day:02d}" for day in range(1, 15)]
e = (-149, 20, -60, 60)
url = "https://gibs.earthdata.nasa.gov/wmts/epsg3857/best/VIIRS_SNPP_CorrectedReflectance_TrueColor/default/{date}/GoogleMapsCompatible_Level9/{{z}}/{{y}}/{{x}}.jpg"

f = "output/nasa.gif"
animate(stack(url.format(date=d), e) for d in dates).save(f, writer="pillow")

Markdown(f"![NASA GIF]({f})")
```
