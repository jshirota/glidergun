```python
from glidergun import search

item = search("sentinel-2-l2a", (-81.1, 46.4, -80.9, 46.5), cloud_cover_percent=5.0)[0]

B8A = item.get("B8A")
B04 = item.get("B04")

ndvi = (B8A - B04) / (B8A + B04)

ndvi.kmeans_cluster(3).color("nipy_spectral_r")
```
