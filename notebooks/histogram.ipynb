```python
from glidergun import grid

g = grid("esri").resample_by(4)


def show(name, g):
    print(name)
    display(g)
    display(g.hist())


show("original", g)

show("negated", -g)

show("set_nan", g.set_nan(g > 200))

show("percent_clip", g.percent_clip(10, 90))

show("slice", g.slice(10))

show("stretch", g.stretch(100, 200))

show("reclass", g.reclass((0, 50, 0), (50, 100, 1), (100, 150, 2), (150, 255, 3)))

show("kmeans_cluster", g.kmeans_cluster(10))

show("cap_range", g.cap_range(50, 120))

show("cap_range_set_nan", g.cap_range(50, 120, set_nan=True))
```
