```python
from IPython.display import Markdown

from glidergun import animate, grid


def tick(g):
    count = g.focal_sum() - g
    return (g == 1) & (count == 2) | (count == 3)


g = grid((120, 80)).randomize() < 0.5

f = "output/game_of_life.gif"
animate(g := tick(g) for _ in range(400)).save(f, writer="pillow", fps=50)

Markdown(f"![Game of Life GIF]({f})")
```
