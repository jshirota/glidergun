```python
from glidergun import grid

points = [
    (-120, 40, 100),
    (-120, 50, 200),
    (-90, 45, 300),
]

point_grid = grid(points, (-140, 30, -50, 60), 4326, 1)
interpolation = point_grid.interp_idw()

point = (-112, 45)
print(f"Estimated value at {point}:", interpolation.value_at(*point))

point_grid, interpolation
```

```python
from glidergun import grid, stack

e = (-140, 30, -50, 60)

basemap = stack("microsoft", e).mean()

points = [
    (-120, 40, 100),
    (-120, 50, 200),
    (-90, 45, 300),
]

point_grid = grid(points, e, 4326, 1)


def overlay(g):
    return basemap.overlay(g.project(3857).color("gist_ncar").to_stack(), 0.3)


(
    overlay(point_grid.distance()),
    overlay(point_grid.interp_idw()),
    overlay(point_grid.interp_clough_tocher()),
    overlay(point_grid.interp_linear()),
    overlay(point_grid.interp_rbf()),
)
```
